```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
t=np.array([5,7,12,16,20]).reshape(-1,1)
m=np.array([40,120,180,210,240])
```

```python
#create a linear regression mod1
LR=LinearRegression()
#fit mod1 to data
LR.fit(t, m)
```

```python
value=LR.predict([[5.5]])
print(value)
#plot the data
plt.scatter(t,m,color='black')
#plot linear regression line
y_pred=LR.predict(t)#t=5,7,12,16,20
print(y_pred)
plt.plot(5.5,y_pred,color='blue',linewidth=3,marker="H")
plt.xlabel('Time')
plt.ylabel('Mass')
plt.title('Linear Regression')
plt.show()
```
